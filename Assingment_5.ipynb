```python
#Regression Analysis for check important parameter who affecting the Precipitation(Rain)
```

```python
#Define Independent Variables columns
IV=Data[['Temperature_Max(Â°F)', 'Temperature_Avg', 'Temperature_Min',
       'pm25', 'pm10', 'o3', 'no2', 'so2', ' co', 'Humidity_Max(%)',
       'Humidity_AVg(%)', 'Humidity_Min(%)', 'Wind_Speed_Max(mph)',
       'Wind_Speed_Avgmph)', 'Wind_Speed_Min(mph)']]
```

```python
#define Dependent column for Pedictive analysis 

DV=Data['Precipitation (in)']
```

```python
#checking data of dependent variable column
DV
```

```python
#IMPORT library of sklearn and module feature_selection
from sklearn.feature_selection import mutual_info_regression
```

```python
#Score Analysis on dependent variable due independent Columns
scores = mutual_info_regression(IV,DV)
```

```python
# check column name inside the Independent variable 
IV.columns
```

```python
#display scores of feature  information regretion
scores
```

```python
# change matrix to pandas series
scores_ser = pd.Series(scores)
```

```python
scores_ser
```

```python
scores_ser.index
```

```python
scores_ser.index = IV.columns
```

```python
scores_ser
```

```python
scores_ser.sort_values(ascending=False)
```

```python
scores_ser.sort_values(ascending=False).plot(kind = "bar")
plt.title(' Features Predicting Precipitation')
```
